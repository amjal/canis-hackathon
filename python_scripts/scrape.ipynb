```python
import wikipedia as wp
import pandas as pd

wiki_df = pd.DataFrame(columns=['Name (English)', 'url', 'content', 'images', 'links', 'categories',
                                'references', 'summary', 'title'])

i = 0
for diplomat in temp['Name (English)'][i:]:
    i += 1
    query = diplomat + " (diplomat)"
    try:
        page = wp.page(query)
        print(i, query, page.title)
        if query.startswith(page.title):
            wiki_df.loc[len(wiki_df)] = [diplomat, page.url, page.content, page.images, page.links, page.categories,
                                    page.references, page.summary, page.title]
            wiki_df.to_csv('../data/wiki.csv', mode='a', header=False, index=False)
    except Exception as e:
        print(e)
        continue
```

```python
wiki_lite = wiki_df[['Name (English)', 'title', 'links', 'categories']]
wiki_lite.to_csv('../data/wiki_lite.csv', index=False)
```

```python
import pandas as pd
import ast

wiki_df = pd.read_csv('../data/wiki_lite.csv')

network_data = []

for index, row in wiki_df.iterrows():
    try:
        links = ast.literal_eval(row['links'])
    except ValueError:
        continue

    for link in links:
        network_data.append((row['title'], link))

wiki_link_network = pd.DataFrame(network_data, columns=['title', 'link'])
wiki_link_network.to_csv('../data/wiki_link_network.csv', index=False)
```
